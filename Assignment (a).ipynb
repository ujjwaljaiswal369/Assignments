```python
'''Q1. Explain the difference between simple linear regression and multiple linear regression. Provide an example of each.'''
```

```python
#Ans.
Linear regression is a supervised machine learning algorithms. The purpose of linear regression is to find best fit line
between the features of dataset. So that we can predict the predicted points with the help of best fit line. Linear 
regression algorithm have three types :
    
1. Simple Linear Regression
2. Multiple Linear Regression
3. Polynomial Linear Regression


Difference between Simple linear regression and Multiple linear regression:
    
1. Simple Linear Regression : simple linear regression purpose is to find linear best fit line between two feature. One
                              is independent feature and another is dependent feature. The best fit line is give predicted
                              points which conncect to the actual datapoints. The best fit line in linear regression
                              should passing through maximum actual datapoints in dataset. 
                             The summition of differences between actual datapoints and predicted points should be minimal
                              There are some equation of simple linear regression but important and most used equation is
                              (hθ(x) = θ0 + θ1x1) in which θ0 is denoted the intercept of best fit line and θ1 is denoted 
                              about the slope and coefficient of the best fit line. 
                              We can get right best fit line by changing the θ value. The cost function is defines the 
                            simple linear regression well, the cost function is also called as mean squared error function
                              When the value of cost function found be minimal by changing the theta value then we got
                              best fit line.
                            
    ex. Let a dataset have two features one is age second is weight. We prepare model by training dataset as like age 
        is input feature and weight is output feature, If we enter the age then it will predicted weight. 
                            
            
2. Multiple linear Regression : Multiple linear regression is a another type of linear regression algorithms. In multiple
                                linear regression, there are have more than two features in dataset or When the dataset 
                                have more than two features then multiple linear regression concept come. 
                                The equation of multiple linear is just like simple linear regression but we should have
                                to add some values also based on number of independent features like:
                                hθ(x) = θ0 + θ1x1 + θ2x3 + θ3x3 + .... + θnxn
                        
    ex. Let a dataset have four feature Gender, No. of rooms, Location and Price. In which Gender , No. of rooms and 
        location helps to predict the price feature after train the model through training dataset.
   

Main differences :
    
1. Simple linear regression is basically applied for that dataset which have total two features whereas Multiple linear 
   regression is applied for more than two features. Both regression have one dependent feature only. 

2. The bestfitline of simple linear regression is linear or straight line whereas Multiple linear regression have plane,
   it may be 2d plane or 3d plane depends upon the no. of independent features having in dataset.

3. The equation of best fit line of simple linear regression is hθ(x) = θ0 + θ1x1 whereas the equation in case of multiple
   linear regression is hθ(x) = θ0 + θ1x1 + θ2x3 + θ3x3 + .... + θnxn. 

4. Simple linear regression have only one slope for one feature whereas the multiple linear regression have multiple 
   slopes of multiple features
```

```python
'''Q2. Discuss the assumptions of linear regression. How can you check whether these assumptions hold in a given dataset?'''
```

```python
#Ans.
Linear regression is a simplest algorithm of machine learning. The machine learning algorithm are helps to make best model
with accuracy. In the theory of linear regression, There are many crucial assumption which we should to know before 
modeling the data.


These are 6 mainly assumptions with their checking method :-


1. Linear Model :
    
This linear model assumptions says that the relationship between independent feature and dependent feature should be 
linear. Their best fit line should be linear. If in case, These two feature have no in relationship then the model will 
not perform accuratly.

Checking = We can check the linear relationship between independent feature and dependent feature by finding best fit line
           in graph. And best fit line should be linear or straight line.
    
    
2. No multicolinearity in dataset :
    
The multicolinearity term means is the features of dataset are highly correlated with each other. When the two or more 
features are seems very similar or their information are similar then it will considered as multicolinearity in dataset. 
If this multicolinearity is increases then the model will getting more complexity and make interfere for models for 
understanding new patterns or information. We should avoid this types of problem in dataset.

Checking = We can check either the multicolinearity presence in dataset or not by using the seaborn library by scattering 
           plot or heatmap function.
    
    
3. Homoscedasticity in dataset :
    
Homoscendicity term means that the spread of residual datapoints have homogeneous or equal spaces between them from linear
regression model. If the spread of residual data is heteroscendicity then it will unsatisfactory model. 

Checking = We can check either dataset have homoscedasticity or heteroscedasticity of residual by seaborn library by 
           scattering of plots.
```

```python
4. No autocorrelation in residuals

One of the critical assumptions of multiple linear regression is that there should be no autocorrelaiton in data. When
the residual are depending on each other than it is known as autocorrelation in data. 
The best example of autocorrelation is stock price where the price of stock is dependent on other stock price.

Checking : We can easily check autocorrelation in residuals by scattering of plots through seaborn librabry


5. No. of observation is greater than no. of predictors

This assumption of linear regression is says that the no. of observation or training dataset should be greater than 
test dataset or predict data. This is very helps in performance of dataset. In linear regression model , for making model
for best performance then the no. of training dataset always should be greater than test dataset.

checking : Checking by analysis and comparison of no. of training dataset or observation and no. of test dataset or 
           prediction dataset.
    
    
6. Predicted values should be normally distributed :
    
This assumption is ensure that the predicted values are should be distributed normally. At the end of model training, 
we should check the predicted values are normally distributed or not.

Checking : We can easily check the distribution of predicted values by applying the plotting density , KDE, QQ plot,etc
```

```python
'''Q6. Explain the concept of multicollinearity in multiple linear regression. How can you detect and address this issue?
'''
```

```python
#Ans.
Multicollinearity is a statistical phenomenon which describe about the correlation between two or more independent 
features of the dataset or when two variables are highly correlated with each other then this situation is called as 
multicollinearity in data which creates the problem in accuracy of the model. Multicollinearity is not good for the 
algorithms, it interfere the model and algorithm accuracy and performance. 

In multiple linear linear regression , there are more than one independent features. There are possibility of existence
of multicollinearity between the variables. We can solve the multicollinearity problem by averaging the similar variables
and replace the average value of variables with the actual value.

We can detect this issue of multicollinearity in variabels by using such method of VIF factor or Variance Inflation Factor
we should calculate the vif of each independent variables and if the value of variable is come more than 1.5 than it is 
considered as presence of multicollinearity in variable.
```

```python
'''Q7. Describe the polynomial regression model. How is it different from linear regression'''
```

```python
#Ans.
Polynomial regression model is a type of supervised machine learning model which is created by using the polynomial 
regression algorithm concept. Polynomial regression algorithm is like simple linear regression algorithm but the 
difference is that linear regression is used where the relation between variables are in linear whereas polynomial 
regression algorithm is used where the relationship between the varibales are not in linear form, there datapoints are
scattered differently, in this type of situation it is hard to construct best fit line and simple linear regression 
algorithm is not worked and we used polynomial regression algorithm. In case of polynomial regression, the best fit
line is bend due to the non linear data

Polynomial regression algorithm are two types one is simple polynomial regression and second is multiple polynomial 
regression. Both the types of polynomial regression is decide by the polynomial degress. The equation of the simple 
polynomial and multiple regression is based upon the value of polynomial degress. If the polynomial degress is 0 then the
equation of simple polynomial regression is [ hθ(x) = θ0 + x to the power 0 ] = constant.
If the polynomial degress is 1 then the equation becomes simple linear regression [ hθ(x) = θ0 + θ1x1 ] .
If the polynomial degress is 2 then the equation becomes [ hθ(x) = θ0 + θ1x1 + θ2x2 ]  and so on ...

And in case of multiple linear regression the equation is if the polynomial degress is 2 then 
[ hθ(x) = θ0 + θ1x1 + θ2x2 + θ3x3 + θ4x1 to the power 2 + θ5x2 to the power 2 + θ6x3 to the power 2 ]
```
