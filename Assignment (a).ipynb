```python
'''Q1. Describe the decision tree classifier algorithm and how it works to make predictions.'''
```

```python
#Ans.

Decision Tree algorithm is a algorithm of supervised machine learning. This algorithm is used for predict the outputs of 
dataset by tree method after creating model. Decision tree algorithm have two types : 
    1. Decision tree Classifier algorithm
    2. Decision tree Regression algorithm
    
Decision tree classifier algorithm are used for those dataset which predictions or output is in categorical form. This 
algorithm is used for classification problem statement. Further, there are two types of classifier algorithm :
    
    This types just tell the type of class splits of decision tree.
    1. ID3 (Iterative Dichotomiser 3) -  The decision tree which have multiclass splits then it will come in category of 
                                         ID3 type.
    2. CART (Classification and Regression Tree) - If the decision tree which have binary class split structure then it 
                                                   will called as CART category.
        
        
The decision tree is made from some specific methods. There have different methods for classification problem and 
continuous problem to make decision tree. For classifier method, the feature is taken which information gain value is more
than other features in dataset then split it based on it have category in feature and find measure of impurity. We can
find measure of impurity by two methods in classifier first is Entropy and another is Ginni impurity.
It is recommended that when dataset is huge then Gini impurity is preferred and when dataset is small size then entropy
impurity is preferred.

Sometimes, the decision tree is made very large which may occur some problem which is overfitting and time complexity
These two problems is created when the decision tree is huge. The overfitting here means is the training accuracy is too
high and the test data accuracy is very low which indicate the overfitting of the model. For get rid from this problem,
We use two techniques first is Post prunning and second is Pre prunning which helps in reducing overfitting. The post
prunning is prunn the branches of trees or make short tree by cutting unnecessary branches. The difference between post 
and pre prunning is, In post prunning , The decision tree is made first, after that analyze the tree then cut the 
unnecessary branch by max_depth parameter. Same operation we do in pre prunning also but in pre prunning we cut the 
brachnes before making the decision tree by hypertunning method.
    
    
How it make predictions : 
    In Desicion Tree, Decision tree is made of nodes. There are mainly three types condidered Root node, Child node and 
    leaf node. Root node is node from which decision tree is start to split the information. Child node is made after the
    root node and leaf node is that node which have no further splits.
    Decision tree took information from dataset and fill in nodes in a specific way. It compute lot of calculation before
    making decision tree. For making Decision tree classifier , It compute lots of calculation like entropy (measure of 
    impurity), information gain, impurity of root node , impurity of child node,etc.
```

```python
'''Q2. Provide a step-by-step explanation of the mathematical intuition behind decision tree classification.'''
```

```python
#Ans.
In decision tree classification, For making proper decision tree we have to do proper mathmatical calculations. The 
dataset output should be categorical. There is different method for decision tree regressor. The process of calculation
are same for binary class and multiclass but there is some difference in case of binary class.
We should first select the feature from the dataset for the root node and split the root node to get parent node and doing
further split to get child node and repeat this process until we get a leaf node.
At the stage of child node we need to check the measure of impurity of child node, for further split. We can check 
impurity by two technique first is entropy and second is ginni impurity. For small dataset we should use entropy and for 
huge data we should use ginni entropy. The entropy is calculated by a specific formula which is:
    
    H(s) = - Pc1*logc1 - Pc2*logc2
    where, c1 and c2 is a category of child node(ex. Yes=c1, No=c2)
    
and after calculating the some values we have to compute the information gain also which tell us about which feature is 
need to put in root node. The formula of information gain is :
    
    Gain(S,f) = H(S) - summition i=1 to n [(|s|/|S|)*H(s)]
         where, H(S) = Entropy of root node
                H(s) = Entropy of child node
                |S| = total number of class or catgoey in root node
                |s| = total number of class or category in child node.
                
                
Step 1 : Select the X feature of dataset as a root node and check the category how many having it (ex. Outlook)
Step 2 : Based on no. of category split root node in child node with same no. of category. (Outlook = Sunny,Overcast,Rain)
Step 3 : Check the every child node how many they have classes of output feature (ex.Sunny(2 yes/3 no),Rain(3 Yes/2 No),
                                                                                  Overcast(4 yes/ 0 no).
Step 4 : Now, Measure the purity of child node by entropy method of each child node the formula for entropy is :
         H(s) = - Pc1*logc1 - Pc2*logc2
         where, c1 and c2 is a category of child node(ex. Yes=c1, No=c2) 
Step 5 : Also calculate the total no. of classes in each node.
Step 6 : Now , Using all the calculated value find the information gain which formula is :
                                                                                  
         Gain(S,f) = H(S) - summition i = 1 to n [(|s|/|S|)*H(s)]
         where, H(S) = Entropy of root node
                H(s) = Entropy of child node
                |S| = total number of class or catgoey in root node
                |s| = total number of class or category in child node.
Step 7 : After selecting the best feature, repeat the process and get the leaf node. 
```

```python
#Ans. 
The geometric concept of decision tree classification is simple but the process of putting information and making its 
branches for constructing decision tree is not easy. There is very logic is used for making a proper decision tree. 

The geometric diagram of any classification problem is made up of nodes. There are four types of nodes : Root node,
parent node, Child node and leaf node.

The root node is main node of the entire decision tree it is basic of all the node, the starting of decision tree making 
is starting from here. When the root node is start spliting then it generate the parent node.

The no. of parent node may be binary or multiclass it based upon the no. category have the feature and after this we have
to check the impurity of parent node for further splits, if the node is pure split then there is no further splits and if
node is impure then there is further splits. The further splits of parent node gives child node.

The child node is also treat same like parent node, we have also find its impurity for its further splits. The splitting 
of all nodes is happening based on information they stored in node and their need for spliting. The process of splitting 
and making bracnches is run continuously until we not get the leaf node. Leaf node is end node or also called as pure 
split because it have not further splits or branches and here the decision tree is end.
```

```python
'''Q5. Define the confusion matrix and describe how it can be used to evaluate the performance of a classification model.'''
```

```python
#Ans.

## Definition
Confusion matrix is a performance matrix which helps in computing the accuracy of models. It is used after the model is 
prepared to test the accuracy of model either the model is performing good or bad. It gives the accuracy of the model in
percentage form which is easy understandable for user about their model. The confusion matrix is used for binary 
classification problem only.


## how it evaluate the performance of model
Confusion matrix is 2x2 matrix which have four block and Each block having some definition such as TP(True positive), 
FP(False Positive), FN(False Negative), TN(True Negative) and the axis is denoted by 0 and 1. The formula for calculating
the accuracy of model is :
    
    Confusion matrix = (TP+TN)/(TP+FP+FN+TN)
    
    The TP,FP,FN,TN is calculated by observing the no. of right or wrong prediction from model outputs. The TP and TN is 
    indicates right predictions and FP and FN is indicates wrong predictions. The reason behind why these matrix called 
    confusion matrix because its parameter is very confusing to understand.
    
The confusion matrix is base of other matrix also like precision matrix, recall matrix and F1 score because these matrix
are take help from confusion matrix for their computing calculations for models. Confusion matrix takes the 0 and 1 from 
actual output(y) and predicting output(y^) from model and compute the no. of wrong and right predictions and fill block
from TP,FP,FN,TN. It is considered that if accuracy of any model come above the 90% then it is good but below 90% means 
to improve the model properly for proper prediction.
```

```python
'''Q6. Provide an example of a confusion matrix and explain how precision, recall, and F1 score can be calculated from it.'''
```

```python
#Ans.
When the model is ready for prediction then for cheking its quality and accuracy , we use performance metrices , There 
are many matrices which shows the accuracy of the models. Out of which, confusion matrice, Recall metrice, precision 
metrice and F1 beta score metrice are very useful. The confusion matrice is a 2x2 matrice which have 4 block and predict
the model by specific formula :
    
    Confusion matrix = (TP+TN)/(TP+FP+FN+TN)
    
    
## Example of confusion metrix
Let suppose a logistic model have following actual data(y) and predicted data(y^) :
________________   
S.No    y   y^ |
 1.     0   0  |
 2.     1   1  |
 3.     0   1  |
 4.     1   0  |
 5.     1   1  |
 6.     1   1  |
_______________|

Then from confusion matrix we count the total no. of right predictions and wrong prediction from confusion metrics and we
get the value of TP=2, FP=1, FN=1, TN=2. Now apply the formula of confusion matric.
Confusion matrix = (TP+TN)/(TP+FP+FN+TN)
                 = (2+2)/(2+1+1+2)
                 = (4)/(6)
                 = 66 %
        
Hence we got the 66% accuracy which indicates the model need improvement.
```

```python
## Explanation of Precision, Recall and F1 beta score :
The precision,recall,F beta score metrices are also used for calculating accuracy but there are some condition for them,
When the dataset is imbalanced then the accuracy of that model cannot calculate directly. Therefore, for preventing this
problem precision metrice,Recall metrice, F1 beta score metrice is used.


1. Precision Metrice : Precision metrice is metrice which evaulate the accuracy of models. Precision metrice is defines as 
                    TP divide by TP plus and FP. 
    
                    Precision = TP/TP+FP
        
                    This metrice is used in case of FP problem for evaluating accuracy , We can find which model should 
                    apply for FP or not by logic. A case is come in model, when model created blunder or give inaccurate 
                    predictions then this type of case is very serious and we should act on FP(False positive).
                    
                    Ex. The model is created of Detecting the spamming mail there are some case occur in predicting spam
                        mail:
                    1. Data  = mail is spam
                       Prediction = mail is spam  (Good)
                        
                    2. Data  = mail is not spam
                       Prediction = mail is not spam  (Good)
                        
                    3. Data  = mail is not spam
                       Prediction = mail is spam  (Blunder)
                        
                        
2. Recall : Recall is also a metric like precision but here the difference is of the FN, Recall is defines as the number
            of true positives divided by no. of true positives plus no. of false positives.
     
            Recall = TP/TP+FN
            
3. F1 Beta score : F Beta score is a combination of precesion and recall metrices, a situation in which the precesion 
                   and recall metrices both participate then it is called F1 beta score. It compute the how many times 
                   a model compute the correct predictions.
                   
                   F beta score = (1+B**2) * [(Precision*Recall) / B**2(Precision+Recall)]
                
                   F beta score have some conditions :
                    
                   1. If the precision and recall are both important then we take Beta value is 1
                   2. If the precision and recall are less important then we take Beta value is 0.5
                   3. If the precision and recall are not important then we take Beta value is 2
                
                   ex. The crashnes of Stock market is best example of F1 beta score because it contain both types of
                            problem precesion and recall.
```
