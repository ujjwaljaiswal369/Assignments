```python
#Ans.

# Data_Encoding
Data Encoding is technique of data preprocessing which encode the categorical feature in numerical feature by different
different methods, for this there are mainly three types of data encoding:
    1. Nominal/OHE (One Hot Encoding)
    2. Ordinal or Label encoding
    3. Target guided ordinal encoding.
    
    
# Uses_in_Data_Science
Data encoding is a part of feature engineering whenever we try to build a machine learning model then in the data
preprocessing step we use data encoding if the given dataset have the categorical columns. Data encoding is helps in converting the categorical feature to 
numerical feature. Because for ML model or DL model a numerical feature is needed only. Because the model only capture 
the numbers not alphabets or other. 
```

```python
'''Q2. What is nominal encoding? Provide an example of how you would use it in a real-world scenario.'''
```

```python
#Ans.

# Nominal_Encoding
Nominal encoding is a type of data encoding technique which is used to convert the categorical feature to numerical 
feature. There are another name of nomial encoding is One hot encoding, the nomial encoding is creating the sparse matrix
of the particular categorical feature. It decide the number of category in categorical column and encode the categories in
binary vector in sparse matrix form and creating a separate coloumn for each category in categorical column. This 
encoding technique is generally used for those data which have no intrinsic order needed or not a specific order needed 
into numerical values. 

# Disadvantages
There are some major disadvantage also of nomial or one hot encoder is, 
1. It creates a sparse matrix which leads to overfitting.
2. It increases the number of columns which reduce the performance of the model.

# use_in_real-world
When we work for real data science industry then there are will be lots of important real world projects like Healthcare
dataset, employee dataset, tips dataset, etc. We should use one hot encoder for those dataset which categorcial columns 
not need specific order of unique values in categorical coulmn then in this case we can use one hot encoder technique. We
can perform Onehotencoder through amazing sklearn library which is easy to use.
```

```python
'''Q3. In what situations is nominal encoding preferred over one-hot encoding? Provide a practical example.'''
```

```python
#Ans.

Nominal encoding is a encoding technique of data encoding and the One Hot encoder is type of nomial encoding. When the 
dataset is nominal type or in another words when the categorical column in dataset have category not in specific order 
then nominal encoding tehcnique is used and for nomial encoding we can use one hot encoder which transform the category 
of categorical column in binary numerical values 0 and 1 in a sparse matrix form .
```

```python
# Practical Example

#importing the libraries
import pandas as pd
from sklearn.preprocessing import OneHotEncoder

#dataset
Tridev = ['Mahadev', 'Narayan', 'Brahma', 'Narayan', 'Mahadev']
df=pd.DataFrame(data=Tridev, columns=['Tridev'])
df
```

```python
# As we know the tridevs are equal , they all three are equal so here there are not any specific order needed all unique
# categories are equal so we can perform for this One  hot encoder.
```

```python
# initialising
encoder = OneHotEncoder()
```

```python
# fitting the data and converting to list
encoded=encoder.fit_transform(df).toarray()
encoded
```

```python
# save in dataframe
encoded=pd.DataFrame(encoded, columns=encoder.get_feature_names_out())
```

```python
# concat the original and encoded data
pd.concat([df, encoded], axis=1)
```

```python
'''Q7.You are working on a project that involves predicting customer churn for a telecommunications company. You have a 
      dataset with 5 features, including the customer's gender, age, contract type, monthly charges, and tenure. Which 
      encoding technique(s) would you use to transform the categorical data into numerical data? Provide a step-by-step 
      explanation of how you would implement the encoding.'''
```

```python
#Ans.

We have a dataset having 30 rows and 5 rows Gender, Age, Contract, Monthly Charges and Tenure. Out of which 2 are 
categorical columns and 3 are numerical columns, the categorical columns are gender and contract, the gender column have
2 unqiue value (Female, Male) and contract column have 3 unique value (Month-to-month, one year, two year). 

We are using one hot encoding because there are not any specific order needed in gender column and contract column. The 
one hot encoder is a part of nominal encoding, it convert the categorical column in binary digits 0 and 1 numerical values.
```

```python
# importing pandas
import pandas as pd
```

```python
# reading dataset
df=pd.read_excel('Customer churn telecommuniaction.xlsx')
df=df.drop('Unnamed: 4', axis=1)
df
```

```python
#  unique values of each column
df.nunique()
```

```python
# Gender have 2 unqie values and Contract column have 3 unique values we can perform one hot encoding because there are 
# not any specific order needed
```

```python
# applying one hot encoder, import onehotencoder through sklearn.preprocessing
from sklearn.preprocessing import OneHotEncoder
```

```python
# initialising
encoder = OneHotEncoder()
```

```python
# fit and transform the column
encoded=encoder.fit_transform(df[['Gender', 'Contract']]).toarray()
```

```python
# encoded output or columns of each unique value
df_encoded=pd.DataFrame(encoded, columns=encoder.get_feature_names_out())
df_encoded
```

```python
# concating the dataframes
df1=pd.concat([df, df_encoded], axis=1)
df1
```
