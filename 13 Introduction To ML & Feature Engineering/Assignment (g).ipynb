```python
'''Q1. Pearson correlation coefficient is a measure of the linear relationship between two variables. Suppose you have 
       collected data on the amount of time students spend studying for an exam and their final exam scores. Calculate 
       the Pearson correlation coefficient between these two variables and interpret the result.'''
```

```python
#Ans.

# importing pandas
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Dataset
df=pd.DataFrame({'Studying/hrs' : [1.10,0.45,2.40,2.10,2,1,3.6,1.0,1.20,3], 
             'Final exam score/100' : [72,69,90,47,76,71,88,40,64,38]})
df
```

```python
# Applying Pearson Correlation Coefficient
df.corr('pearson')
```

```python
# Explanation
We have a dataset of the student studying for the exam and the final exam score. We use the pearson correlation 
coefficient for knowing the relationship between these two variables. The pearson correlation coefficient is helps to
determine the relationship of two variables by specified in a range of -1 to +1 for linear relationship of variables.
After applying pearson correlation, The variable Studying/hrs is 15% correlate with the other variable Final exam 
score/100. If 15% up down of one variable can also change 15% of other variable.
```

```python
'''Q2. Spearman's rank correlation is a measure of the monotonic relationship between two variables. Suppose you have 
       collected data on the amount of sleep individuals get each night and their overall job satisfaction level on a 
       scale of 1 to 10. Calculate the Spearman's rank correlation between these two variables and interpret the result.'''
```

```python
#Ans.
import pandas as pd
```

```python
df=pd.DataFrame({'Sleep(hrs)' : [8,6,7,8,7,6,9,10,6,7],
             'Job Satisfaction' : [8,7,8,7,9,5,10,10,4,8]})
df
```

```python
df.corr('spearman')
```

```python
#interpreting the relation
By using the spearman correlation, we get the 79% relationship between the variables if the sleeping hours increase then 
job satisfaction also increases with 79%. The reason behind using spearman correlation is that because these variables 
is creating monotonic relationship with each other. 

As we think with simple logic that if a employee is sleeps more it means he get more free hours to spend himself , it 
means he have less office work hour and he is enjoying his life more. The more sleeping hours a employee sleeps means he 
get more free hours from the job which means he/she is more satisfied from his job.
```

t_2.png](attachment:780e4533-92cf-4569-9ff0-e0e5940d275c.png)

```python
#Ans.

# importing pandas
import pandas as pd
```

```python
# dataframe
df=pd.DataFrame({'Age(Years)' : [25,42,37,19,31,28],
             'Soft Drink Preference' : ['Coke','Pepsi', 'Mountain dew', 'Coke', 'Pepsi', 'Coke']})
df
```

```python
# Onehotencoder importing
from sklearn.preprocessing import OneHotEncoder
```

```python
# initialising
encoder = OneHotEncoder()
```

```python
# fit and transform the data
encoded = encoder.fit_transform(df[['Soft Drink Preference']]).toarray()
```

```python
# save in dataframe
df1=pd.DataFrame(encoded, columns= encoder.get_feature_names_out())
```

```python
# concatinate the both data
df=pd.concat([df,df1], axis=1)
```

```python
df
```

```python
# Pearson correlation
df.corr('pearson')
```

```python
By using pearson correlation matric we can observe that with increasing age the human preferred for pepsi most and coke
is preferred lowest. 
```
