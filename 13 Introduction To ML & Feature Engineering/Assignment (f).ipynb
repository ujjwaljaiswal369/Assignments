```python
'''Q1. What is the difference between Ordinal Encoding and Label Encoding? Provide an example of when you might choose 
       one over the other.'''
```

```python
#Ans.
Ordinal Encoding = Ordinal encoding is a another encoding technique which is mainly used for when a categorical column 
                   needs encoding with sequence of categories. The ordinal encoding preserve the order of the category in
                   categorical columns. 
        
Suppose a dataset have a column of categorical column which have a category of Good, Better and Best and for encoding
we needs to encode the column with in sequence of Good, Better and Best to 0, 1, 2 respectively. We can perform the 
ordinal encoding from the sklearn library easily.
```

```python
## Example

# making dataframe
import pandas as pd
df = pd.DataFrame({'Product' : ['Earphone', 'Headset', 'Micrphone', 'TV', 'Mobile'], 
                  'Quality' : ['Good', 'Best', 'Better', 'Best', 'Good']}
)
```

```python
df
```

```python
# importing ordinalencoder library from sklearn.preprocessing
from sklearn.preprocessing import OrdinalEncoder
```

```python
# initialising
encoder = OrdinalEncoder(categories=[['Good', 'Better', 'Best']])
```

```python
# fit and transform 
encoded_quality=encoder.fit_transform(df[['Quality']])
encoded_quality
```

```python
# store encoded column in dataframe
df['Quality']=pd.DataFrame(encoded_quality, columns=['encoded_quality'])
```

```python
# final_result
df
```

```python
# Label Encoding 
Label Encoding = Label encoding is also encode the categorical column but it is not for sequence categories, it is not 
                 like the ordinal encoding. It encode the categories in categorical column alphabetically . Label encoding
                 label the unique values with the numbers like 0,1,2 in alphabetically order.
```

```python
# making dataframe
import pandas as pd
df = pd.DataFrame({'Product' : ['Earphone', 'Headset', 'Micrphone', 'TV', 'Mobile'], 
                  'Quality' : ['Good', 'Best', 'Better', 'Best', 'Good']}
)
```

```python
df
```

```python
# importing labelencoder library
from sklearn.preprocessing import LabelEncoder
```

```python
# initialising
encoder = LabelEncoder()
```

```python
# fit and transform 
encoded_quality = encoder.fit_transform(df['Quality'])
encoded_quality
```

```python
# store encoded column in dataframe
df['Quality'] = pd.DataFrame(encoded_quality, columns=['encoded_quality'])
```

```python
# final_result
df
```

```python
'''Q2. Explain how Target Guided Ordinal Encoding works and provide an example of when you might use it in a machine 
       learning project.'''
```

```python
#Ans.
Target Guided Ordinal Encoding = Target Guided Ordinal Encoding is a technique of data encoding which encode the 
                                 categorical values to numerical values by replacing from the mean and median values of 
the target variable. Target guided ordinal encoding applying for those datasets which categorical feature have strong 
relationship with the target varible. If the target variable have outliers then we can use median and otherwise we can 
use mean values for encoding.

Target guided ordinal encoding is establish the monotonic relationship with the target variable which helps in improving 
the predictive power of the model.
```

```python
## Example

Suppose we have tips dataset which have 244 rows and 7 columns, Out of which 3 are numerical and 4 are categorical 
columns and we considered the total bill column as a target variable and encode the day and time coulmn because we can
notice the target variable total_bill have strong correlation with the day and time feature.

# importing libraries
import seaborn as sns
import pandas as pd
```

```python
# dataframe
df=sns.load_dataset('tips')
df
```

```python
# mean finding of total bill with respect to day
mean_day=df.groupby('day')['total_bill'].mean().to_dict()
mean_day
```

```python
# mean finding of total bill with respect to time
mean_time=df.groupby('time')['total_bill'].mean().to_dict()
mean_time
```

```python
# mapping the encoded values
df['encoded_day']=df['day'].map(mean_day)
```

```python
df
```

```python
df['encoded_time'] = df['time'].map(mean_time)
```

```python
# dropping the previous columns 
df=df.drop(['day', 'time'], axis=1)
```

```python
# final result
df.head()
```

```python
'''Q3. Define covariance and explain why it is important in statistical analysis. How is covariance calculated?'''
```

```python
#Ans.

# Definition_and_explanation_of_Covariance
Covariance = Covariance is a statistical analysis of the random variables, it tells the measurement of the relation 
             between two random variables. The spreadness of the data of one variable from the mean position is known as 
             variance and if the spreadness of the data is shown of the two variables then it is known as co-variance.
             Variance is shown spreadness of one random variable and co-variance is shown spreadness of two random variable
             
             Let suppose there are two random variables x and y then the formula of finding the covariance is : 
                Cov(x,y) = (xi - x^-)(yi - y^-)/n-1
                
                where, xi  = value of each x datapoints
                       yi  = value of each y datapoints
                       x^- = x bar/sample mean of x variable
                       y^- = y bar/sample mean of y variable
                       n   = no. of datapoints
                        
             The value of covariance can come positive and negative. The positive covariance shows the alignment of the 
             both x and y variables, if the value of x variable is increase then value of y variable is also increases.
             and same happen in case of decrement also.
             The negative covariance is shows the vice versa relationship of random variables. If the x is increase or
             decrease then y is also decrese or increase respectively.
            
# Use of covariance            
The covariance technique of knowing the relationship between the random variables is very useful in building the machine
learning model. The covariance and correlarion helps in finding the best features for predicting model by showing the 
relation between the random variables and the target variables. 

# Advantage_and_Disadvantages_of_covariance
There are some advantage and disadvantage also of covariance: 
    
1. The covariance is tell the relationship between the two random variables (Advantage)
2. It is unable to tell the covariance value in limit which results there is hard to understand the relation of variables
(Disadvantage) 

There are some methods which helps to get rid from its disadvantage. We can get the value of covariance in range of -1 to
+1 by using two major technique which is:
    
1. pearson correlation coefficient
2. Spearnmen correlation 

- The formula of pearson correlation coefficient is, divide the covariance of the random variable by standard deviation
   of x and y variable. It is denoted by row:  ρ(x,y) = cov(x,y)/σx.σy 
    
- The formula of spearmen correlation is also same but there is substitution of rank x and rank y variable in place of
    x and y.  ρ(R(x),R(y)) = cov(R(x),R(y))/σR(x).σR(y) 
```

```python
#Ans.

# importing libraries
import pandas as pd
import warnings
warnings.filterwarnings('ignore')
```

```python
# Dataset
df=pd.DataFrame({'Age' : [21,32,18,23,25,41,19], 
              'Income' : [25000,65000,18000,27000,12000,37000,22000],
              'Education level' : ['Bsc Math', 'B.Tech', '12th pass', 'Bsc Math', 'BCA', 'B.Tech', 'B.Tech']
              })
df
```

```python
# Encoding the categorical column from LabelEncoder
from sklearn.preprocessing import OrdinalEncoder

encoder = OrdinalEncoder(categories=[['12th pass', 'Bsc Math', 'BCA', 'B.Tech']])
df['Education level']=encoder.fit_transform(df[['Education level']])
```

```python
df
```

```python
# Covariance
df.cov()
```

```python
# Pearson correlation coefficient
df.corr(method='spearman')
```

```python
# Explanation
We have a dataset of Age,Income and Education level dataset. We first the encode the categorical feature 'Education level'
after that we apply the correlation and covariance. For getting the limit in a range ,we use pearson correlation
coefficient. 

The correlation matrix tells the relation between all the feature, The relation with Age and Income feature shows the
60% relation which mean if the 60% age increase then income is also increase and Relation with the education level is 
shown negative correlation 61% which means the increment of 61% in age can decrease the 61% of education level.

As well as same relation observation we can find with education level to age and income and Income to Age and education 
level.
```

```python
'''Q6. You are working on a machine learning project with a dataset containing several categorical variables, including
       "Gender" (Male/Female), "Education Level" (High School/Bachelor's/Master's/PhD),and "Employment Status" 
       (Unemployed/Part-Time/Full-Time). Which encoding method would you use for each variable, and why?'''
```

```python
#Ans.

# importing pandas
import pandas as pd
```

```python
# dataset 
df=pd.DataFrame({'Gender' : ['Male', 'Female', 'Male', 'Male', 'Female', 'Male'],
             'Education Level' : ['High School', 'Master.s', 'PhD', 'High School', 'Bachelor.s',  'PhD'],
             'Employment Status' : ['Unemployed', 'Full-Time', 'Full-Time', 'Part-Time', 'Unemployed', 'Full-Time']})
df
```

```python
# using OneHotEncoder for Gender and Employment Status features
from sklearn.preprocessing import OneHotEncoder
```

```python
# initialising
encoder = OneHotEncoder()
```

```python
# fit and transform
encoded = encoder.fit_transform(df[['Gender', 'Employment Status']])
```

```python
# save encoded values in dataframe
encoded_df = pd.DataFrame(encoded.toarray(), columns=encoder.get_feature_names_out())
```

```python
encoded_df
```

```python
# dropping useless feature
df=df.drop(['Gender', 'Employment Status'], axis=1)
```

```python
df
```

```python
# concatinate the original df and encoded df
df=pd.concat([df, encoded_df], axis=1 )
```

```python
df
```

```python
# Using OrdinalEncoder for Employment Level feature
from sklearn.preprocessing import OrdinalEncoder
```

```python
# initialising and setting the order of categories
encoder = OrdinalEncoder(categories=[['High School', 'Bachelor.s', 'Master.s', 'PhD']])
```

```python
# fit and transform the feature
encoded = encoder.fit_transform(df[['Education Level']])
```

```python
# save enocded values in dataframe
df['Education Level'] = pd.DataFrame(encoded, columns= encoder.get_feature_names_out())
```

```python
# final result
df
```

```python
We have a dataset which have 6 rows and 3 columns. Out of which all are categorcial features. According to the nature of
categories in features the different different encoding technique is used for encoding the features. 

For Gender and Employment status feature we use Onehotencoder technique because the categories present in these features 
are not needed in specific order so we apply one hot encoder and For Education level feature we apply the ordinal 
encoding technique because there is a specific order needed of high school, bachelor, master, phd categories . So for a 
specific order requirement we use the Ordinal encoding technique.
```
